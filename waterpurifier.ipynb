```python
driver = webdriver.Chrome()
```

```python
data_list = []

for page_url in allpages:
    # Navigate to the current page
    driver.get(page_url)
    time.sleep(3)  # Wait for the page to load
    
    # Load the page content into BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    rodata = soup.find_all('div', class_="yKfJKb row")
    
    for d in rodata:
        RO_Name_ = d.find('div', class_="KzDlHZ")
        price_ = d.find('div', class_="Nx9bqj _4b5DiR")
        original_price_ = d.find('div', class_="yRaY8j ZYYwLA")
        rating_ = d.find('div', class_="XQDdHH")
        
        # Extract text if element is found, otherwise set to None
        RO_Name_ = RO_Name_.text if RO_Name_ else None
        price_ = price_.text[1:] if price_ else None
        original_price_ = original_price_.text[1:] if original_price_ else None
        rating_ = rating_.text if rating_ else None
        
        # Extract features
        features_ = [feature.text for feature in d.find_all('li')]
        
        # Create a dictionary for the current item
        item_dict = {
            'RO_Name': RO_Name_,
            'price': price_,
            'original_price': original_price_,
            'rating': rating_,
            'features_1': features_
        }
        
        # Append the item dictionary to the data list
        data_list.append(item_dict)

# Close the WebDriver


# Process the data_list as needed
print(data_list)
```

```python
import pandas as pd
df_Ro = pd.DataFrame(data_list)
df_Ro
```
